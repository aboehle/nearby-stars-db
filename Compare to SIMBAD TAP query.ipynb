## Compare the two tables

```python
cmd = '''
SELECT stars_simbad.main_id, 1/(stars_simbad.plx*1e-3)
FROM stars_simbad
    LEFT OUTER JOIN stars_simbadtap ON stars_simbad.main_id = stars_simbadtap.main_id
WHERE stars_simbadtap.oid IS NULL
ORDER BY stars_simbad.plx DESC
'''
c.execute(cmd)
rows = c.fetchall()
print 'stars in regular query and not in tap query:',len(rows)
print
for row in rows:
    print row
```

```python
cmd = '''
SELECT stars_simbadtap.main_id
FROM stars_simbadtap
    LEFT OUTER JOIN stars_simbad ON stars_simbad.main_id = stars_simbadtap.main_id
WHERE stars_simbad.main_id IS NULL
ORDER BY stars_simbadtap.plx_value DESC
'''
c.execute(cmd)
rows = c.fetchall()
print 'stars in tap query and not in regular query:',len(rows)
print
for row in rows[0:10]:
    print row
```

These are stars for which the main type of double/multiple star, but the companion is a planet!
